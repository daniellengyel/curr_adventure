```python
def createU(max_row, dim):
    I = np.eye(dim)
    I[:, 0] = np.ones(dim)
    Q, R = np.linalg.qr(I)
    Q = Q.T
    row_tmp = Q[max_row].copy()
    Q[max_row] = Q[0]
    Q[0] = row_tmp
    return jnp.array(Q)
    

def get_a(D_min, sum_sqrt, dim):
    disc = 8 * dim * D_min + sum_sqrt**2 + 8 * D_min
    return jnp.sqrt(sum_sqrt**2 / (2 * dim * D_min) + sum_sqrt * jnp.sqrt(disc) / (2 * dim * D_min) + 1/dim + 1)

def get_lambda_max(D_min, sum_sqrt, dim):
    disc = 8 * dim * D_min + sum_sqrt**2 + 8 * D_min
    numerator = sum_sqrt * jnp.sqrt(disc) + 4 * dim * D_min + sum_sqrt**2 + 4 * D_min
    denom = D_min**(1.5) * 2 * jnp.sqrt(2/dim) * jnp.sqrt(sum_sqrt * jnp.sqrt(disc) + 2 * dim * D_min + sum_sqrt**2 + 2 * D_min)
    return numerator/denom

def get_full_sing_vals(D, dim):
    D_diag = jnp.abs(jnp.diag(D))
    max_row = jnp.argmin(D_diag)
    sum_sqrt = jnp.sum(jnp.sqrt(D_diag)) - jnp.sqrt(D_diag[max_row])
    lmbda_max = get_lambda_max(jnp.min(D_diag), sum_sqrt, dim)
    a = get_a(jnp.min(D_diag), sum_sqrt, dim)
    lmbda = np.array(jnp.sqrt(lmbda_max / (2 * a * D_diag)))
    lmbda[max_row] = lmbda_max
    lmbda = jnp.array(lmbda)
    sing_val = jnp.sqrt(lmbda)
    return sing_val

def get_S(D, dim):
    D_diag = jnp.abs(jnp.diag(D))
    max_row = jnp.argmin(D_diag)
    
    sing_val = get_full_sing_vals(D, dim)
    U = createU(max_row, dim)
#     print(repr(U))
#     print(sing_val)
    return jnp.diag(sing_val) @ U

# get_a(jnp.sum(jnp.sqrt(jnp.linspace(1, 1000, 5))) - 1, 5)
# get_lambda_max(jnp.sum(jnp.sqrt(jnp.linspace(1, 1000, 5))) - 1, 5)

# dim = 20
# D = jnp.diag(jnp.linspace(1, 10, dim))


# # print(get_S(D, dim))

# # S = get_S(D, dim)
# # print(S)
# # jnp.diag(S.T @ D @ S)
```

```python


def helper_linesearch(f, g, c1, c2):

    def helper(X, search_direction):
        f0 = f(X)
        f1 = g(X)
        dg = jnp.inner(search_direction, f1)

        def armijo_rule(alpha):
            return f(X + alpha * search_direction) > f0 + c1*alpha*dg
        
        def armijo_update(alpha):
            return c2*alpha
            
        alpha = 1
        while armijo_rule(alpha):
            alpha = armijo_update(alpha)


        return alpha

    return helper

def loss_getter(dim, N, H, sig):
    sing_vals = jnp.diag(get_full_sing_vals(H, dim))
    def helper(X):

        U = X.reshape(N, dim).T
        S = sing_vals @ U
        
        S_inv = jnp.linalg.inv(S)
        
        first_term = S_inv.T @ jnp.diag(S.T @ H @ S)
        second_term = jnp.linalg.norm(S_inv, ord="fro")**2
        third_term = S_inv.T @ jnp.ones(dim)
        third_term = jnp.linalg.norm(third_term)**2
        
        return jnp.linalg.norm(first_term)**2 + sig**2 * (second_term + third_term)

    return helper

# def loss_getter(dim, N, H, sig):
#     sing_vals = jnp.diag(get_full_sing_vals(H, dim))
#     def helper(X):

#         S = X.reshape(N, dim).T
        
#         S_inv = jnp.linalg.inv(S)
        
#         first_term = S_inv.T @ jnp.diag(S.T @ H @ S)
#         second_term = jnp.linalg.norm(S_inv, ord="fro")**2
#         third_term = S_inv.T @ jnp.ones(dim)
#         third_term = jnp.linalg.norm(third_term)**2
        
#         return jnp.linalg.norm(first_term)**2 + sig**2 * (second_term + third_term)

#     return helper

def optimize_uncentered_S(H, sig, max_steps=15, jrandom_key=None, x_curr=None):
    
    dim = H.shape[0]
    N = dim
    
    H = (H + H.T) / 2. # to combat numerical inaccuracies. 
    D, U = jnp.linalg.eig(H)
    U = jnp.real(U)

    # print("D", repr(jnp.real(D)))
    D = jnp.real(jnp.diag(D))

    if x_curr is None:
        x_curr = createU(4, dim).flatten()
#         if jrandom_key is None:
#             x_curr = jnp.diag(jnp.sqrt(sig / jnp.abs(jnp.diag(D)))).flatten()
#             pass
#         else:
#             x_curr = jrandom.normal(jrandom_key, shape=(dim * dim,)) * 0.1
        

    
    l = loss_getter(dim, N, D, sig)
    g_l = grad(l)
    h_l = jacfwd(g_l)
    linesearch = helper_linesearch(l, g_l, c1=0.01, c2=0.9)

    eps = 1e-20

    res = []
    res_x = []

    for t in tqdm(range(max_steps)):

        
        res.append(l(x_curr))
#         res_x.append(U.dot(x_curr.reshape(N, dim).T)) # TURN ON FOR HISTORY 

        curr_grad = g_l(x_curr)
    
        # print(curr_grad)
#         curr_hess = h_l(x_curr) + 0.01*jnp.eye(5**2)
#         search_direction = -jnp.linalg.inv(curr_hess) @ curr_grad #+ np.random.normal(size=(25))*0.1
        search_direction = -curr_grad + np.random.normal(size=(25))*0.1

        if jnp.linalg.norm(curr_grad) < eps:
            break    

        alpha = linesearch(x_curr, search_direction)

        S = x_curr.reshape(N, dim).T

        SS = S.dot(S.T)

        x_curr += alpha*search_direction 
    
    return U.dot(x_curr.reshape(N, dim).T), res
```

```python
def createU(dim):
    I = np.eye(dim)
    I[:, 0] = np.ones(dim)
    Q, R = np.linalg.qr(I)
    Q = Q.T
    return jnp.array(Q)

def orthog_linesearch(l, c1, c2):

    def helper(X, search_direction, A):
        f0 = l(X)
        g_tau_0 = -1/2 * jnp.linalg.norm(A, "fro")**2
        
        def armijo_rule(alpha):
            return (l(search_direction(alpha)) > f0 + c1*alpha*g_tau_0) 
        
        def armijo_update(alpha):
            return c2*alpha
            
        alpha = 10
        while armijo_rule(alpha):
            alpha = armijo_update(alpha)


        return alpha

    return helper


def optimize_U(c, num_iter):
    """Constraint is U.c = \bar{c} 1 and U.U^T = I"""
    
    dim = len(c)
    
    # init X
    X = jnp.eye(dim)
    I = jnp.eye(dim)
    
    X = X.flatten()
#     X = jnp.array([[-0.26554573,  0.62542372,  0.43805724,  0.40086329,
#                0.43098156],
#              [-0.3254421 ,  0.64933193, -0.29632447, -0.36856546,
#               -0.49880536],
#              [-0.36026767, -0.26370683,  0.76995586, -0.29417584,
#               -0.34827355],
#              [-0.55210503, -0.26383148, -0.24347228,  0.67330204,
#               -0.33609314],
#              [-0.6236667 , -0.21923701, -0.26112618, -0.4044664 ,
#                0.57549547]]).flatten()
    
    def l(U):
        U_matrix = U.reshape(dim, dim)
        
        Uc = U_matrix**2 @ c 
        return jnp.linalg.norm(Uc - jnp.ones(dim) * jnp.mean(c)) + jnp.exp(jnp.linalg.norm(U_matrix[0] - 1/jnp.sqrt(dim) * jnp.ones(dim)))

    
    g_l = grad(l)
    linesearch = orthog_linesearch(l, c1=0.1, c2=0.5)
    eps = 1e-9
    
    l_hist = []
    for _ in tqdm(range(num_iter)):

        num_iter -= 1
        
        G = g_l(X).reshape(dim, dim)
        l_hist.append(l(X))
#         print(jnp.linalg.norm(G))
        
        if jnp.linalg.norm(G) < eps:
            break
        
        A = G @ X.reshape(dim, dim).T - X.reshape(dim, dim) @ G.T
        
        Y = lambda tau: jnp.linalg.inv(I + tau/2 * A) @ (I - tau/2 * A) @ X.reshape(dim, dim)
        
        alpha = linesearch(X, Y, A)
#         print(alpha)
        
        X = Y(alpha)
        
    return X, l_hist
        
    
U, l_hist = optimize_U(jnp.array([1, 2, 3, 4, 100]), 10) 

plt.plot(l_hist)
```
