# 1 D examples 

```python
jrandom_key = jrandom.PRNGKey(0)
D = jnp.array([[1]])

sig = 0.1
jrandom_key, subkey = jrandom.split(jrandom_key)

F = Quadratic(1/2 * D, jnp.zeros(1), 0, sig=sig)
```

```python
X = jnp.linspace(-1.5, 1, 51)
Y = []
Y_true = []
for x in X:
    jrandom_key, subkey = jrandom.split(jrandom_key)
    Y.append(F.f(jnp.array([x]), subkey))
    Y_true.append(F.f(jnp.array([x])))



plt.plot(X, Y)
plt.plot(X, Y_true, color="lightblue")

plt.savefig("noisy", dpi=500)

```

```python
X[10]
```

```python
X_grad = jnp.linspace(-1.5, -0.5, 50)
fg = F.f1(jnp.array([-1.]))
Y_grad = [F.f(jnp.array([-1.])) + fg * (x + 1) for x in X_grad]

plt.plot(X, Y)
plt.scatter(jnp.array([-1.]), F.f(jnp.array([-1.])), color='orange')
plt.plot(X_grad, Y_grad)
plt.plot(X, Y_true, color="lightblue")

plt.savefig("true_grad", dpi=500)

```

```python
X_1_short = -1
X_1_short_idx = 10
X_2_short_idx = 13
X_2_short = X[X_2_short_idx]

fg_grad_short = (Y[X_2_short_idx] - Y[X_1_short_idx])/(X_2_short - X_1_short)

X_grad_short = jnp.linspace(-1.1, X_2_short + 0.1)
Y_grad_short = [Y[X_1_short_idx] + fg_grad_short * (x + 1) for x in X_grad_short]

plt.plot(X, Y)
plt.scatter(jnp.array([-1.]), F.f(jnp.array([-1.])), color='orange')
plt.scatter(jnp.array([X_2_short]), Y[X_2_short_idx], color='green')
plt.scatter(jnp.array([X_1_short]), Y[X_1_short_idx], color='green')



plt.plot(X_grad, Y_grad)
plt.plot(X_grad_short, Y_grad_short)
plt.plot(X, Y_true, color="lightblue")
plt.savefig("short_grad", dpi=500)

plt.show()
```

```python
X_1_long = -1
X_1_long_idx = 10
X_2_long_idx = 40
X_2_long = X[X_2_long_idx]

X_3_long_idx = 41
X_3_long = X[X_3_long_idx]



fg_grad_long = (Y[X_2_long_idx] - Y[X_1_long_idx])/(X_2_long - X_1_long)

X_grad_long = jnp.linspace(-1.1, X_2_long + 0.1)
Y_grad_long = [Y[X_1_long_idx] + fg_grad_long * (x + 1) for x in X_grad_long]

plt.plot(X, Y)
plt.scatter(jnp.array([-1.]), F.f(jnp.array([-1.])), color='orange')
plt.scatter(jnp.array([X_2_long]), Y[X_2_long_idx], color='green')
plt.scatter(jnp.array([X_1_short]), Y[X_1_short_idx], color='green')



plt.plot(X_grad, Y_grad)
plt.plot(X_grad_long, Y_grad_long)
plt.plot(X, Y_true, color="lightblue")

plt.savefig("long_grad", dpi=500)





```
