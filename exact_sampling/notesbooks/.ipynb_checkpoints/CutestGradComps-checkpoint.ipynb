```python
def get_mc_loss(F, x_0, num_runs):
    g_F = F.f1(x_0)
    H = F.f2(x_0)
    def helper(grad_getter, jrandom_key):
        errs = []
        for _ in range(num_runs):
            jrandom_key, subkey = jrandom.split(jrandom_key)
            sgd = grad_getter.grad(F, x_0, subkey, H=H)[0]
            errs.append(float(jnp.linalg.norm(sgd - g_F)**2))
        errs = jnp.array(errs)
        return jnp.mean(errs), jnp.std(errs), jnp.percentile(errs, 25), jnp.percentile(errs, 75)
    return helper
```

```python

def lambda_coeff_zero_set_max_lambda(D_diag, sig, l_max):
    dim = len(D_diag)

    if dim == 1:
        return jnp.array([min([l_max, 4 * sig / jnp.abs(D_diag)[0]])])

    if sum(D_diag) < 0:
        D_diag = -D_diag

    if all(D_diag > 0):
        lmbda_no_max = lambda_no_max_h(D_diag, sig)
        if all(lmbda_no_max <= l_max):
            return lmbda_no_max

    D_arg_sorted = jnp.argsort(D_diag)

    n_upper = len(D_diag) - sum(D_diag <= 0)
    n_lower = 0
    n_curr = (n_upper + n_lower)//2
    print(n_curr)
    
    def lmbda_helper(num_not_h):
        K1 = jnp.sum(jnp.sqrt(D_diag[D_arg_sorted[num_not_h:]]))
        K2 = jnp.sum(D_diag[D_arg_sorted[:-num_not_h]])
        
        c1 = sig * jnp.sqrt(dim * l_max / 2) * K1
        c2 = l_max * K2
        
        disc = 81 * c1**2 - 12 * c2**3 
        if disc < 0:
            r = jnp.sqrt(12 * c2**3)
            theta = jnp.arccos(9 * c1 / r)
            mu1 = jnp.cos(theta/3)
            a = mu1**2 * c2 * 4/3.
            
        else:
            
            cbrt = jnp.cbrt(jnp.sqrt(disc) + 9 * c1)
            a_sqrt = jnp.cbrt(2/3) * c2 / cbrt + cbrt/(jnp.cbrt(2 * 9))
        
            a = a_sqrt**2


        lmbda = np.ones(dim) * float(l_max) 
        lmbda[np.array(D_arg_sorted[num_not_h:])] = sig * np.sqrt(dim * l_max / (2 * a * D_diag[D_arg_sorted[num_not_h:]]))

        return lmbda
    
    counter = 0
    while True:

        lmbda = lmbda_helper(n_curr)
            
        if lmbda is None:

            raise Exception("lmbda was None for D_diag {} and l_max {}".format(repr(D_diag), l_max))
            break


        if any(lmbda > l_max):
            n_upper = n_curr
        else:
            n_lower = n_curr

        n_curr = (n_upper + n_lower) // 2

        counter += 1
        if n_upper == n_lower:
            break
    
    return jnp.array(lmbda)
```
